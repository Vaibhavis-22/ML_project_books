```python
from pyspark.sql.functions import col, sum as Fsum

df.select([Fsum(col(c).isNull().cast("int")).alias(c) for c in df.columns]).show()
```

```python
from pyspark.ml.recommendation import ALS



# Train ALS model
als = ALS(
    userCol="user",
    itemCol="book_id",
    ratingCol="rating",
    coldStartStrategy="drop",
    implicitPrefs=False,
    nonnegative=True,
    rank=10,
    maxIter=10,
    regParam=0.1
)

```

```python

als_model = als.fit(df)
```

```python
als_model.save("models/als_model")
```
